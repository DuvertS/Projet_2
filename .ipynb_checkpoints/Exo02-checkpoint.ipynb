## Aperçu du projet

Ce projet a pour objectif d’aider une entreprise à identifier les types de films à produire pour son nouveau studio de cinéma. À travers une analyse des tendances actuelles des films populaires, nous proposons des recommandations stratégiques basées sur les données de films (genres, durée, notes, votes).


### Importation des Données Fimls : Analyse Préliminaire

```python
# Inportation des bibliothèques

import sqlite3
import pandas as pd

# Connexion à la base SQLite
conn = sqlite3.connect("im.db")

# Charger les données nécessaires (films avec votes > 100)
query = """
SELECT 
    mb.movie_id,
    mb.primary_title AS title,
    mb.start_year AS year,
    mb.runtime_minutes AS duration_min,
    mb.genres,
    mr.averagerating,
    mr.numvotes
FROM 
    movie_basics mb
JOIN 
    movie_ratings mr
ON 
    mb.movie_id = mr.movie_id
WHERE 
    mb.start_year IS NOT NULL
    AND mr.numvotes > 100
"""
df = pd.read_sql_query(query, conn)
df.head()

```

### Nettoyage et Préparation des Données

```python
# Supprimer les valeurs manquantes
df.dropna(inplace=True)

# Convertir les types de colonnes
df["year"] = df["year"].astype(int)
df["duration_min"] = df["duration_min"].astype(float)

# Extraire le genre principal (le 1er dans la liste)
df["main_genre"] = df["genres"].apply(lambda x: x.split(",")[0] if "," in x else x)

df.head()
```

### Visualisation : Répartition des Genres Principaux

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Comptage des genres principaux
plt.figure(figsize=(10, 6))
top_genres = df["main_genre"].value_counts().head(10)
sns.barplot(x=top_genres.values, y=top_genres.index, hue=top_genres.index, palette="crest", legend=False)
plt.title("Top 10 des genres de films")
plt.xlabel("Nombre de films")
plt.ylabel("Genre")
plt.show()
```

### Visualisation : Distribution des Notes Moyennes IMDb

```python
# La distribution des notes moyennes des films
plt.figure(figsize=(8, 5))
sns.histplot(df["averagerating"], bins=20, kde=True, color="skyblue")
plt.title("Distribution des notes moyennes IMDb")
plt.xlabel("Note")
plt.ylabel("Nombre de films")
plt.show()
```

### Visualisation : Durée des Films Selon Leur Genre

```python
# La durée des films selon leur genre
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=df[df['main_genre'].isin(top_genres.index)],
    x='main_genre', y='duration_min', hue='main_genre',
    palette="Set2", legend=False
)
plt.title("Profit des films par genre")
plt.xlabel("Genre")
plt.ylabel("duration (min)")
plt.xticks(rotation=45)
plt.show()
```

### Analyse de Corrélation : Notes, Votes et Durée des Films

```python
# La corrélation entre les notes, les votes et la durée des films
plt.figure(figsize=(6, 4))
corr = df[['averagerating', 'numvotes', 'duration_min']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Corrélation entre note, votes et durée")
plt.show()
```

### Fimls a Faible Rentabilité Potentielle

```python
# Films avec faible rentabilité potentielle (note < 6 et peu de votes)
low_success = df[(df["averagerating"] < 6) & (df["numvotes"] < 500)]
print(f"Nombre de films à faible succès potentiel : {len(low_success)}")

# Répartition par genre
plt.figure(figsize=(8, 5))
sns.countplot(
    data=low_success,
    y="main_genre",
    hue="main_genre",
    order=low_success["main_genre"].value_counts().index,
    palette="flare",
    legend=False
)
plt.show()
```

### Recommendation 
 Produire des films dans les genres : Drame, Biographie, Thriller 
Viser une durée entre 90 et 150 minutes 
Maintenir une qualité élevée (note IMDb > 7)
 
 Éviter :Films trop longs (>180 min)Films trop spécifiques ou expérimentaux avec peu de votes

### Résumé des conclusions

1. **Les genres dominants** chez les films bien notés et populaires sont :
   - Drame
   - Biographie
   - Thriller

2. **Durée optimale** : Les films qui réussissent le mieux durent entre **90 et 150 minutes**.

3. **Éviter les films** :
   - Trop longs (>180 minutes)
   - Trop expérimentaux ou peu populaires
   - Avec des genres rares ou mal classés
